#### **Calculate mean, standard deviation, minimum, and maximum**

```python
def calculate_basic_stats(data, label=""):
    """Calculate mean, std, min, max."""
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # Sample std
    min_val = np.min(data)
    max_val = np.max(data)
    
    print(f"\n{label} AUC:")
    print(f"  Data: {data}")
    print(f"  Mean: {mean:.4f}")
    print(f"  Std:  {std:.4f}")
    print(f"  Min:  {min_val:.4f}")
    print(f"  Max:  {max_val:.4f}")
    print(f"  Range: [{min_val:.4f}, {max_val:.4f}]")
    
    return {
        'mean': mean,
        'std': std,
        'min': min_val,
        'max': max_val,
        'data': data
    }
```

---

#### **Calculate 95% confidence interval -- Following paper: resampling of the data in each set**

```python
def bootstrap_ci(data, n_bootstrap=1000, confidence=0.95):
    """
    Calculate bootstrap confidence interval.
    Following paper: resampling of the data in each set.
    """
    n = len(data)
    bootstrap_means = []
    
    print(f"\n  Performing bootstrap with {n_bootstrap} resamples...")
    
    np.random.seed(42)  # For reproducibility
    for _ in range(n_bootstrap):
        # Resample with replacement
        resample = np.random.choice(data, size=n, replace=True)
        bootstrap_means.append(np.mean(resample))
    
    bootstrap_means = np.array(bootstrap_means)
    
    # Calculate percentile CI
    alpha = 1 - confidence
    lower_percentile = (alpha / 2) * 100
    upper_percentile = (1 - alpha / 2) * 100
    
    ci_lower = np.percentile(bootstrap_means, lower_percentile)
    ci_upper = np.percentile(bootstrap_means, upper_percentile)
    
    print(f"  Bootstrap mean: {np.mean(bootstrap_means):.4f}")
    print(f"  Bootstrap std:  {np.std(bootstrap_means):.4f}")
    print(f"  95% CI: [{ci_lower:.4f}, {ci_upper:.4f}]")
    
    return ci_lower, ci_upper
```

#### **Calculate statistics**

```python
# Calculate statistics
print("\n" + "="*80)
print("CALCULATING STATISTICS")
print("="*80)

train_stats = calculate_basic_stats(train_aucs, "Training")
val_stats = calculate_basic_stats(val_aucs, "Validation")
test_stats = calculate_basic_stats(test_aucs, "Test")

# Calculate uncertainty metric based on approach
print("\n" + "="*80)
print("UNCERTAINTY QUANTIFICATION")
print("="*80)

if approach == "5-fold CV":
    print("\nFor 5-Fold CV: Using RANGE of values (as per paper)")
    print("Note: Bootstrap CI is NOT used for 5-fold CV in the paper")
    
    uncertainty_method = "range"
    
elif approach == "cohort split":
    print("\nFor Cohort Split: Using Bootstrap 95% CI (as per paper)")
    print(f"Bootstrap parameters: {n_bootstrap} resamples")
    
    uncertainty_method = "bootstrap"
    
    print("\nBootstrap CI for Training AUC:")
    train_ci_lower, train_ci_upper = bootstrap_ci(train_aucs, n_bootstrap)
    
    print("\nBootstrap CI for Validation AUC:")
    val_ci_lower, val_ci_upper = bootstrap_ci(val_aucs, n_bootstrap)
    
    print("\nBootstrap CI for Test AUC:")
    test_ci_lower, test_ci_upper = bootstrap_ci(test_aucs, n_bootstrap)
```
