```python
import pandas as pd

print("="*80)
print("EVENT RATE CALCULATION FOR TABLE 1")
print("="*80)

# ========== CANADIAN DATASET (n=298) ==========
print("\n1. HEAD-NECK-PET-CT [Canadian Dataset]")
print("-" * 80)

canada_df = pd.read_csv("canada.csv")
total_canada = len(canada_df)

# Calculate event counts and percentages
lrf_canada = canada_df['Locoregional'].sum()
dm_canada = canada_df['Distant'].sum()
os_canada = canada_df['Death'].sum()

print(f"Total patients: {total_canada}")
print(f"LRF events: {lrf_canada} ({lrf_canada/total_canada*100:.1f}%)")
print(f"DM events: {dm_canada} ({dm_canada/total_canada*100:.1f}%)")
print(f"OS events: {os_canada} ({os_canada/total_canada*100:.1f}%)")

# ========== DUTCH DATASET (n=137) ==========
print("\n2. HEAD-NECK-RADIOMICS-HN1 [Dutch/Maastro Dataset]")
print("-" * 80)

maastro_df = pd.read_csv("maastro.csv")
total_maastro = len(maastro_df)

# Calculate event counts and percentages
lrf_maastro = maastro_df['event_locoregional_recurrence'].sum()
dm_maastro = maastro_df['event_distant_metastases'].sum()
os_maastro = maastro_df['event_overall_survival'].sum()

print(f"Total patients: {total_maastro}")
print(f"LRF events: {lrf_maastro} ({lrf_maastro/total_maastro*100:.1f}%)")
print(f"DM events: {dm_maastro} ({dm_maastro/total_maastro*100:.1f}%)")
print(f"OS events: {os_maastro} ({os_maastro/total_maastro*100:.1f}%)")

# ========== CMC INDIAN DATASET (n=102) ==========
print("\n3. HNC-RADIOMICS [CMC/Indian Dataset]")
print("-" * 80)

cmc_df = pd.read_csv("cmc.csv")
total_cmc = len(cmc_df)

# Calculate event counts and percentages
lrf_cmc = cmc_df['event_locoregional_recurrence'].sum()

print(f"Total patients: {total_cmc}")
print(f"LRF events: {lrf_cmc} ({lrf_cmc/total_cmc*100:.1f}%)")

# ========== SUMMARY FOR TABLE 1 ==========
print("\n" + "="*80)
print("SUMMARY - COPY THIS TO TABLE 1")
print("="*80)

print(f"\nCanadian (n={total_canada}):")
print(f"  LRF: {lrf_canada} ({lrf_canada/total_canada*100:.1f}%)")
print(f"  DM: {dm_canada} ({dm_canada/total_canada*100:.1f}%)")
print(f"  OS: {os_canada} ({os_canada/total_canada*100:.1f}%)")

print(f"\nDutch (n={total_maastro}):")
print(f"  LRF: {lrf_maastro} ({lrf_maastro/total_maastro*100:.1f}%)")
print(f"  DM: {dm_maastro} ({dm_maastro/total_maastro*100:.1f}%)")
print(f"  OS: {os_maastro} ({os_maastro/total_maastro*100:.1f}%)")

print(f"\nIndian (n={total_cmc}):")
print(f"  LRF: {lrf_cmc} ({lrf_cmc/total_cmc*100:.1f}%)")

print("\n" + "="*80)
```
