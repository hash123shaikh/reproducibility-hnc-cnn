## **`Calculate Median Follow-up Time for CMC Dataset`**

================================================

### This script calculates the median follow-up time for the CMC cohort, separating patients by recurrence status.

**Author:** `Hasan Shaikh`

```python
"""
Calculate Median Follow-up Time for CMC Dataset
================================================
"""

import pandas as pd
import numpy as np

# File path - adjust this to your actual file location
file_path = 'cmc.csv'

print("="*80)
print("CMC DATASET - MEDIAN FOLLOW-UP CALCULATION")
print("="*80)

# Read the dataset
df = pd.read_csv(file_path)

print(f"\nTotal patients in dataset: {len(df)}")
print(f"Patients with recurrence: {(df['event_locoregional_recurrence'] == 1).sum()}")
print(f"Patients without recurrence: {(df['event_locoregional_recurrence'] == 0).sum()}")

# Get follow-up times
follow_up_days = df['locoregional_recurrence_in_days']

# Overall statistics
print("\n" + "="*80)
print("OVERALL FOLLOW-UP (All 102 patients)")
print("="*80)
overall_median_days = follow_up_days.median()
overall_median_years = overall_median_days / 365.25
overall_mean_days = follow_up_days.mean()
overall_mean_years = overall_mean_days / 365.25

print(f"Median follow-up: {overall_median_days:.0f} days = {overall_median_years:.2f} years")
print(f"Mean follow-up:   {overall_mean_days:.1f} days = {overall_mean_years:.2f} years")
print(f"Range: {follow_up_days.min():.0f} - {follow_up_days.max():.0f} days")

# Calculate IQR
q1 = follow_up_days.quantile(0.25)
q3 = follow_up_days.quantile(0.75)
print(f"IQR: {q1/365.25:.1f} - {q3/365.25:.1f} years")

# Separate by recurrence status
recurrence_yes = df[df['event_locoregional_recurrence'] == 1]
recurrence_no = df[df['event_locoregional_recurrence'] == 0]

# Patients WITH recurrence (event = 1)
print("\n" + "="*80)
print(f"PATIENTS WITH RECURRENCE (n={len(recurrence_yes)})")
print("="*80)
time_to_recurrence = recurrence_yes['locoregional_recurrence_in_days']
median_recur_days = time_to_recurrence.median()
median_recur_years = median_recur_days / 365.25
mean_recur_days = time_to_recurrence.mean()
mean_recur_years = mean_recur_days / 365.25

print(f"Median time to recurrence: {median_recur_days:.0f} days = {median_recur_years:.2f} years")
print(f"Mean time to recurrence:   {mean_recur_days:.1f} days = {mean_recur_years:.2f} years")
print(f"Range: {time_to_recurrence.min():.0f} - {time_to_recurrence.max():.0f} days")

# Patients WITHOUT recurrence (event = 0) - censored
print("\n" + "="*80)
print(f"PATIENTS WITHOUT RECURRENCE - Censored (n={len(recurrence_no)})")
print("="*80)
follow_up_censored = recurrence_no['locoregional_recurrence_in_days']
median_cens_days = follow_up_censored.median()
median_cens_years = median_cens_days / 365.25
mean_cens_days = follow_up_censored.mean()
mean_cens_years = mean_cens_days / 365.25

print(f"Median follow-up: {median_cens_days:.0f} days = {median_cens_years:.2f} years")
print(f"Mean follow-up:   {mean_cens_days:.1f} days = {mean_cens_years:.2f} years")
print(f"Range: {follow_up_censored.min():.0f} - {follow_up_censored.max():.0f} days")

# Check 2-year minimum for censored patients
min_censored_days = follow_up_censored.min()
min_censored_years = min_censored_days / 365.25
print(f"\nMinimum follow-up for censored patients: {min_censored_days:.0f} days = {min_censored_years:.2f} years")

below_2yr = (follow_up_censored < 730).sum()
if below_2yr == 0:
    print(f"✓ All {len(recurrence_no)} censored patients meet 2-year minimum follow-up requirement")
else:
    print(f"✗ WARNING: {below_2yr} censored patients have <2 years follow-up")

# Summary for Table 1
print("\n" + "="*80)
print("RECOMMENDATION FOR TABLE 1")
print("="*80)
print(f"Overall median follow-up: {overall_median_years:.2f} years ({overall_median_days:.0f} days)")
print(f"\nFor Table 1, report as:")
print(f"  Option 1: Median {overall_median_years:.1f} years")
print(f"  Option 2: Median {overall_median_years:.2f} years")
print(f"  Option 3: Median {median_cens_years:.1f} years (IQR: {q1/365.25:.1f}-{q3/365.25:.1f})")

# Additional verification: show first few rows
print("\n" + "="*80)
print("SAMPLE DATA (first 10 patients)")
print("="*80)
sample = df[['id', 'event_locoregional_recurrence', 'locoregional_recurrence_in_days']].head(10)
sample['follow_up_years'] = sample['locoregional_recurrence_in_days'] / 365.25
print(sample.to_string(index=False))

print("\n" + "="*80)
print("CALCULATION COMPLETE")
print("="*80)
```
