```python
#!/usr/bin/env python3
"""
Script to check which IDs from CSV are missing in PNG folder
and which PNG files are extra (not in CSV)

Usage:
    python check_id_match.py

Requirements:
    - proceed_radiomics_103_patients(two_years).csv in same directory
    - cmc_c/ folder in same directory
"""

import pandas as pd
import os

def check_id_matches():
    # File paths
    csv_file = "proceed_radiomics_103_patients(two_years).csv"
    folder_path = "cmc_c"
    
    # Check if files exist
    if not os.path.exists(csv_file):
        print(f"ERROR: CSV file '{csv_file}' not found!")
        print(f"Please make sure the file is in the same directory as this script.")
        return
    
    if not os.path.exists(folder_path):
        print(f"ERROR: Folder '{folder_path}' not found!")
        print(f"Please make sure the folder is in the same directory as this script.")
        return
    
    # Read the CSV file
    print(f"Reading CSV file: {csv_file}")
    df = pd.read_csv(csv_file)
    
    # Get the list of IDs from the CSV
    csv_ids = set(df['id'].astype(str))
    print(f"\n{'='*70}")
    print(f"Total IDs in CSV file: {len(csv_ids)}")
    print(f"{'='*70}")
    print(f"Sample IDs: {', '.join(sorted(csv_ids)[:5])}...")
    
    # Get the list of PNG files in the folder
    print(f"\nReading PNG files from: {folder_path}")
    png_files = [f for f in os.listdir(folder_path) if f.endswith('.png')]
    
    # Extract IDs from PNG filenames (remove .png extension)
    png_ids = set([f.replace('.png', '') for f in png_files])
    print(f"\n{'='*70}")
    print(f"Total PNG files in folder: {len(png_files)}")
    print(f"{'='*70}")
    print(f"Sample PNG IDs: {', '.join(sorted(png_ids)[:5])}...")
    
    # Find IDs in CSV but not in folder (MISSING)
    missing_ids = sorted(csv_ids - png_ids)
    print(f"\n{'='*70}")
    print(f"❌ IDs in CSV but MISSING PNG files in folder: {len(missing_ids)}")
    print(f"{'='*70}")
    if missing_ids:
        for idx, id in enumerate(missing_ids, 1):
            print(f"  {idx:3d}. {id}")
        
        # Save to file
        with open('missing_ids.txt', 'w') as f:
            f.write("IDs in CSV but missing PNG files:\n")
            f.write("="*70 + "\n")
            for id in missing_ids:
                f.write(f"{id}\n")
        print(f"\n✓ Saved list to 'missing_ids.txt'")
    else:
        print("  ✓ None! All CSV IDs have corresponding PNG files.")
    
    # Find PNG files not in CSV (EXTRA)
    extra_ids = sorted(png_ids - csv_ids)
    print(f"\n{'='*70}")
    print(f"➕ PNG files in folder but NOT in CSV (EXTRA): {len(extra_ids)}")
    print(f"{'='*70}")
    if extra_ids:
        for idx, id in enumerate(extra_ids, 1):
            print(f"  {idx:3d}. {id}.png")
        
        # Save to file
        with open('extra_pngs.txt', 'w') as f:
            f.write("PNG files in folder but not in CSV:\n")
            f.write("="*70 + "\n")
            for id in extra_ids:
                f.write(f"{id}.png\n")
        print(f"\n✓ Saved list to 'extra_pngs.txt'")
    else:
        print("  ✓ None! All PNG files correspond to CSV IDs.")
    
    # Perfect matches
    matches = sorted(csv_ids & png_ids)
    print(f"\n{'='*70}")
    print(f"✓ Perfect matches (ID in both CSV and folder): {len(matches)}")
    print(f"{'='*70}")
    
    # Summary
    print(f"\n{'='*70}")
    print(f"📊 SUMMARY")
    print(f"{'='*70}")
    print(f"CSV IDs:              {len(csv_ids):4d}")
    print(f"PNG files:            {len(png_files):4d}")
    print(f"Perfect matches:      {len(matches):4d} ✓")
    print(f"Missing from folder:  {len(missing_ids):4d} ❌")
    print(f"Extra in folder:      {len(extra_ids):4d} ➕")
    print(f"{'='*70}")
    
    # Calculate percentages
    if len(csv_ids) > 0:
        match_percentage = (len(matches) / len(csv_ids)) * 100
        print(f"\nMatch rate: {match_percentage:.1f}% of CSV IDs have PNG files")
    
    if len(missing_ids) > 0:
        print(f"\n⚠️  WARNING: {len(missing_ids)} patients in CSV are missing PNG files!")
    
    if len(extra_ids) > 0:
        print(f"\n⚠️  NOTE: {len(extra_ids)} PNG files don't match any CSV ID")

if __name__ == "__main__":
    print("="*70)
    print("ID MATCHING CHECK: CSV vs PNG Files")
    print("="*70)
    print()
    
    try:
        check_id_matches()
    except Exception as e:
        print(f"\n❌ ERROR: {e}")
        import traceback
        traceback.print_exc()
    
    print("\n" + "="*70)
    print("Check complete!")
    print("="*70)
```
